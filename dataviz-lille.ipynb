```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hopital = pd.read_csv('data/localisation-hopitaux.csv', sep=';', index_col=5)
```

```python
hopital
```

```python
hopital["commune"].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
hopital["commune"].value_counts().plot(kind='barh')
plt.ylabel("Qté Hopital", fontsize=9)
plt.xlabel('Commune', fontsize=9)
ax.set_title('Nombre d\'hopital par ville', fontsize=14)
for i in ax.patches:
    ax.text(i.get_width()+.1, i.get_y()+.31, \
            str(round((i.get_width()), 2)), fontsize=9, color='dimgrey')
```

```python
population = pd.read_csv('data/population-par-age-et-familles-monoparentales-dans-les-communes.csv', sep=';')
population.columns = ['code_iris', 'commune', 'nbre_65_et_plus', 'nbre_moins_20', 'nbre_familles_monoparentales', 'nbre_plus_75', 'geom', 'geo_point_2d']
population
```

```python
ax = population[['nbre_65_et_plus', 'nbre_moins_20', 'nbre_familles_monoparentales', 'nbre_plus_75']].plot(kind='barh', stacked=True,
              figsize=(30,80), color=['dodgerblue', 'slategray', 'slateblue', 'coral'], fontsize=10);
plt.ylabel('Commune', fontsize=9)
plt.xlabel('Nombre', fontsize=9)
ax.set_title('Population par tranche d\'age et par commune', fontsize=14)
```
